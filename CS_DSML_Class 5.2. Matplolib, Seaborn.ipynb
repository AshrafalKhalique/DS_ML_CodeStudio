## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## Simple Examples
x=np.arange(0,10)
y=np.arange(11,21)
x,y
```

```python
plt.scatter(x, y) # plotting scatter plot, means only intersected point will be shown
plt.xlabel("X axis") # axis labelling
plt.ylabel("Y axis")
plt.title("Scatter Plot") # plot name
```

```python
y = x**3
y
```

```python
## plt plot
plt.plot(x,y,'y*', linestyle='dashed', linewidth=2, markersize=12) # y* means yellow color, star point
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

```python
## Creating Subplots
plt.subplot(2,2,1) #(nrows, ncolms, index)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'y*--')
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title("sine curve")
plt.plot(x, y)
```

```python
# Bar Plot
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 

plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 

plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  

plt.show()
```

```python
# Histograms
a = np.array([22, 87, 5, 43, 56, 73, 55, 54, 11, 20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
plt.show()
```

```python
# Pie Chart
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0.4, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

```python
# Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips") # default dataset inside seaborn
df.head()
```

```python
df.shape
```

```python
df.corr() # correlation between column and row
```

```python
sns.heatmap(df.corr()) # darker the color - weaker the correlation
```

```python
# Pair Plot
sns.pairplot(df) # correlate between elemts inside column
```

```python
sns.pairplot(df, hue="time")
```

```python
# Checking distribution of the features
sns.distplot(df['tip'])
```

```python
# JoinPlot -> relationship of numerical features
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.countplot(y='sex', data=df)
```

```python
sns.barplot(x='sex', y='total_bill', data=df)
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=df)
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
```
